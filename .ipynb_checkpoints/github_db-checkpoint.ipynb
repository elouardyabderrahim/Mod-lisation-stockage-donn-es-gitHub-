```python
import pandas as pd
import sqlalchemy as sly
import psycopg2
```

```python
contributors_file_path='contributors_one.csv'
general_data_file_path='The_final_translated_data.csv'
```

```python
# Create the engine to connect to the PostgreSQL database
#‘postgresql://username:password@databasehost:port/databasename’
engine = sly.create_engine('postgresql+psycopg2://postgres:admin@localhost:5432/brief-four')
```

```python
general_data=pd.read_csv(general_data_file_path)
contributors_data=pd.read_csv(contributors_file_path)
```

### topics correction

```python
import ast

# Convert the strings back to lists using ast.literal_eval
general_data['topics'] = general_data['topics'].apply(ast.literal_eval)

# Explode the 'topics' column to split the lists into individual elements
df_exploded = general_data.explode('topics')

# Reset the index of the exploded data frame
general_data = df_exploded.reset_index(drop=True)
```

```python
# Write data into the table in PostgreSQL database
general_data.to_sql('githubdata',engine)
```

```python
from sqlalchemy import text

sql = '''
    DROP TABLE githubdata;
'''
with engine.connect() as conn:
    query = conn.execute(text(sql))         
```

### Contributors dataFrame clean and correct columns

```python
import pandas as pd
import ast

# Given data frame
data =contributors_data["contributors"].to_frame()

df = pd.DataFrame(data)

# Function to extract values from the 'contributors' column
# Function to extract values from the 'contributors' column
def extract_contributor_info(row):
    if pd.notnull(row['contributors']):
        contributor_info = ast.literal_eval(row['contributors'])
        return pd.Series(contributor_info)
    else:
        return pd.Series({'src_repository': None, 'contributor': None, 'number_of_contributions': None})

# Applying the function to extract new columns
df[['src_repository', 'contributor', 'number_of_contributions']] = df.apply(extract_contributor_info, axis=1)

# Dropping the original 'contributors' column
df = df.drop(columns=['contributors'])

print(df)
```

### Add Ids to tables for insertion:

```python
# old methode 
df = df.reset_index()
df = df.rename(columns={"index":"contributor_ID"})
df['contributor_ID'] = df.index + 1
```
